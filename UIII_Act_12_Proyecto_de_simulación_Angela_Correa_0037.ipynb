### Ejercicio 14: Joyero 💍
▶ Atributos: Materiales (oro/plata), técnica de engaste, valor de inventario, certificados de autenticidad, diseño propio y tienda.

```python
# Definición de la clase en singular
class joyero:
    # Atributos de la clase
    def __init__(self, materiales, tecnica_engaste, valor_inventario, certificados_autenticidad, diseño_propio, tienda):
        self.materiales = materiales
        self.tecnica_engaste = tecnica_engaste
        self.valor_inventario = valor_inventario
        self.certificados_autenticidad = certificados_autenticidad
        self.diseño_propio = diseño_propio
        self.tienda = tienda

# Lista global para almacenar los objetos de tipo joyero
lista_joyeros = []

# --- Funciones dentro de la clase (simuladas como funciones de gestión del objeto) ---

def agregar_joyero():
    print("\n--- REGISTRAR NUEVO JOYERO ---")
    materiales = input("Materiales (Ej. Oro, Plata): ")
    tecnica_engaste = input("Técnica de engaste (Ej. Pavé, Bisel): ")
    valor_inventario = input("Valor de inventario (Ej. $5000): ")
    certificados_autenticidad = input("¿Tiene certificados de autenticidad? (Sí/No): ")
    diseño_propio = input("¿Es diseño propio? (Sí/No): ")
    tienda = input("Nombre de la tienda: ")

    # Creamos el objeto con la clase joyero
    nuevo_joyero = joyero(materiales, tecnica_engaste, valor_inventario, certificados_autenticidad, diseño_propio, tienda)
    lista_joyeros.append(nuevo_joyero)
    print("\n✨ ¡Joyero agregado con éxito al sistema!")

def listar_joyeros():
    print("\n--- INVENTARIO DE JOYEROS ---")
    if len(lista_joyeros) == 0:
        print("No hay ningún joyero registrado en este momento.")
        return

    # Encabezados de la tabla con formato y alineación
    print(f"{'No.':<4} | {'Materiales':<12} | {'Engaste':<12} | {'Valor':<10} | {'Certificado':<12} | {'Diseño Pro.':<12} | {'Tienda':<15}")
    print("-" * 90)

    # Recorremos la lista y mostramos los atributos alineados
    for indice, j in enumerate(lista_joyeros):
        print(f"{indice + 1:<4} | {j.materiales:<12} | {j.tecnica_engaste:<12} | {j.valor_inventario:<10} | {j.certificados_autenticidad:<12} | {j.diseño_propio:<12} | {j.tienda:<15}")

def actualizar_joyero():
    print("\n--- ACTUALIZAR JOYERO ---")
    if len(lista_joyeros) == 0:
        print("No hay joyeros para actualizar.")
        return

    listar_joyeros()
    posicion = int(input("\nIntroduce el número del joyero que deseas modificar: ")) - 1

    if posicion >= 0 and posicion < len(lista_joyeros):
        print("\n--- Introduce los nuevos datos ---")
        lista_joyeros[posicion].materiales = input("Nuevos Materiales: ")
        lista_joyeros[posicion].tecnica_engaste = input("Nueva Técnica de engaste: ")
        lista_joyeros[posicion].valor_inventario = input("Nuevo Valor de inventario: ")
        lista_joyeros[posicion].certificados_autenticidad = input("¿Nuevos Certificados? (Sí/No): ")
        lista_joyeros[posicion].diseño_propio = input("¿Nuevo Diseño propio? (Sí/No): ")
        lista_joyeros[posicion].tienda = input("Nueva Tienda: ")
        print("\n🔄 ¡Datos del joyero actualizados correctamente!")
    else:
        print("Número de joyero no válido.")

def borrar_joyero():
    print("\n--- ELIMINAR JOYERO ---")
    if len(lista_joyeros) == 0:
        print("No hay joyeros para eliminar.")
        return

    listar_joyeros()
    posicion = int(input("\nIntroduce el número del joyero que deseas borrar: ")) - 1

    if posicion >= 0 and posicion < len(lista_joyeros):
        lista_joyeros.pop(posicion)
        print("\n🗑️ ¡El joyero ha sido eliminado del sistema!")
    else:
        print("Número de joyero no válido.")

# --- Sistema de Inteligencia Artificial Simulada y Menú Interactivo ---
def menu_interactivo():
    print("======================================================")
    print("🤖 BIENVENIDO AL SISTEMA DE IA DE GESTIÓN DE JOYERÍA 🤖")
    print("======================================================")
    print("Analizando tendencias de diseño y optimizando inventario...")

    while True:
        print("\n🤖 [Asistente de IA] ¿Qué acción deseas realizar hoy?")
        print("1. Agregar un nuevo joyero")
        print("2. Listar inventario de joyeros (Ver tabla)")
        print("3. Actualizar datos de un joyero")
        print("4. Borrar un joyero")
        print("5. Salir del sistema")

        opcion = input("Selecciona una opción (1-5): ")

        if opcion == "1":
            agregar_joyero()
        elif opcion == "2":
            listar_joyeros()
        elif opcion == "3":
            actualizar_joyero()
        elif opcion == "4":
            borrar_joyero()
        elif opcion == "5":
            print("\n👋 ¡Gracias por usar el Asistente de IA de Joyería! Cerrando sistema...")
            continuar = False
        else:
            print("Opción no válida. Por favor selecciona una opción del 1 al 5.")

# Ejecución del programa principal
menu_interactivo()
print("Programa realizado por Angela Michelle Correa Lopez NL = 14")
```
